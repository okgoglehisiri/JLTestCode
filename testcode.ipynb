```python
# 
import pandas as pd
import matplotlib.pyplot as plt             #
# 
url='https://www.salesanalytics.co.jp/591h' #URL
table=pd.read_csv(url,                      #URL
                  index_col='Month',        #Month
                  parse_dates=True)         #
# 
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [10, 8]
# 
table.plot()
plt.title('Passengers')                            #
plt.ylabel('Monthly Number of Airline Passengers') #
plt.xlabel('Month')                                #
plt.show()
```
